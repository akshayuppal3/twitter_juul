```python
# check if the string is number.
def is_number(s):
    try:
        int(s)
        return True
    except ValueError:
        return False
# check if the string is date
def is_date(string):
    try: 
        parse(string)
        return True
    except ValueError:
        return False
```

```python
# return the type of node :(user -0,burst-2 or url-1) ... it works to diff user and burst
def check_type_node(s):
    if is_number(s):
        return(0)
    elif is_date(s):
        return(2)
    else:
        return(1)
```

```python
## determing the node potential based on type of node
def get_node_potential(s):
    node_type = check_type_node(s)
    if node_type is 0:
        if int(s) in list(user_potential.userID):
            pot = user_potential[user_potential.userID == int(s)]
            return(pot.prom.values[0],pot.nprom.values[0])
    if node_type is 1: #url
        if s in list(url_potential.url):
            pot = url_potential.loc[url_potential.url == s]
            return(pot.prom.values[0],pot.nprom.values[0])
    if node_type is 2: #burst
        s = datetime.strptime(s,'%Y-%m-%d').strftime('%Y-%m')
        if s in list(burst_potential.date):
            pot = burst_potential.loc[burst_potential.date == s]
            return(pot.prom.values[0],pot.nprom.values[0])
```
